```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
from pathlib import Path

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
csv_file = Path("cities.csv")
city_data_df = pd.read_csv(csv_file)

# Display sample data
city_data_df.head()
```

```python
# Function to find the hotel near given coordinates using the Geoapify API
def find_hotel_near_coordinates(latitude, longitude):
    
    url = 'https://api.geoapify.com/v1/places'

    params = {
        'lat': latitude,
        'lon': longitude,
        'type': 'hotel',
        'radius': 10000,
        'apiKey': geoapify_key
    }

    response = requests.get(url, params=params).json()

    if 'features' in response and len(response['features']) > 0:
        hotel_name = response['features'][0]['properties']['name']
        country = response['features'][0]['properties']['country']
        return hotel_name, country
    else:
        return None, None

# Apply the function to the DataFrame to find the hotels
city_data_df[['Hotel Name', 'Country']] = city_data_df.apply(lambda row: find_hotel_near_coordinates(row['Lat'], row['Lng']), axis=1, result_type='expand')

# Create a map using hvplot scatter plot
map_center = [city_data_df['Lat'].mean(), city_data_df['Lng'].mean()]
city_map = city_data_df.hvplot.scatter(
    x='Lng',
    y='Lat',
    size='Humidity',  # Set the size of the points based on humidity
    hover_cols=['City', 'Country', 'Humidity', 'Hotel Name'],  # Include humidity in hover message
    geo=True,
    tiles='EsriImagery',
    title='City Map with Humidity',
    xlim=(-180, 180),
    ylim=(-90, 90),
    colorbar=True,
    width=600,
    height=400
)

city_map
```
